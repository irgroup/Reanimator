```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder_path = "/workspace/pdfs"
```

```python
file_sizes = []

# Loop through all files in the specified folder
for filename in os.listdir(folder_path):
    # Check if the file is a PDF
    if filename.endswith(".pdf"):
        # Get the size of the file in kilobytes (KB)
        file_path = os.path.join(folder_path, filename)
        file_size_kb = os.path.getsize(file_path) / 1024
        file_sizes.append(file_size_kb)

# Filter out non-positive sizes
file_sizes = [size for size in file_sizes if size > 0]

# Check if there are any positive file sizes to plot
if file_sizes:
    # Determine the minimum and maximum file sizes for the log scale bins
    min_size = min(file_sizes)
    max_size = max(file_sizes)

    # Create more bins with higher resolution in the lower range
    bins = np.logspace(np.log10(min_size), np.log10(max_size), 50)  # Increase bin count

    # Plot the distribution of file sizes with a log-log scale
    plt.figure(figsize=(10, 6))
    plt.hist(file_sizes, bins=bins, edgecolor="black")
    plt.xscale("log")  # Log scale for x-axis
    #plt.yscale("log")  # Log scale for y-axis
    plt.xlabel("File Size (KB)")
    plt.ylabel("Number of Files")
    plt.title("Log Scaled Distribution of PDF File Sizes (High Resolution on X-Axis)")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()
else:
    print("No positive file sizes to plot.")
```

```python
import os
import shutil

# Specify the original folder containing the PDF files
folder_path = "/workspace/pdfs"
# Specify the backup folder where the files below 5 KB will be copied
backup_folder = "/workspace/backup"

# Create the backup folder if it does not exist
os.makedirs(backup_folder, exist_ok=True)

# Loop through all files in the specified folder
for filename in os.listdir(folder_path):
    # Check if the file is a PDF
    if filename.endswith(".pdf"):
        # Get the size of the file in kilobytes (KB)
        file_path = os.path.join(folder_path, filename)
        file_size_kb = os.path.getsize(file_path) / 1024
        
        # Check if the file size is below 5 KB
        if file_size_kb < 15:
            # Copy the file to the backup folder
            backup_path = os.path.join(backup_folder, filename)
            shutil.copy2(file_path, backup_path)
            
            # Delete the file from the original folder
            os.remove(file_path)            
            print(f"Copied and deleted: {filename} ({file_size_kb:.2f} KB)")
```
